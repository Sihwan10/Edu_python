```python
array = [6,5,3,1,8,7,2,4]
#[1,2,3,4,5,6,7,8]
```

```python
n = len(array)
```

```python
print(n)
```

```python
#리스트 버블정렬
for i in range(n-1):
    for j in range(n-i-1):
        #print(j)
        if array[j] > array[j+1]:
            tmp = array[j]
            array[j] = array[j+1]
            array[j+1] = tmp
        print(array)
        
#for i in range(n-1):
    #if array[i] > array[i+1]:
        #tmp = array[i]
        #array[i] = array[i+1]
        #array[i+1] = tmp
    #print(array)
```

```python
#최솟값
array = [6,5,3,1,8,7,2,4]

min = array[0]

for i in array:
    if i < min:
        min = i
print(min)
```

```python
#선택정렬

array = [6,5,3,1,8,7,2,4]

for i in range(n):
    min = i
    for j in range(i,n):
        if array[j] < array[min]:
            min = j
    tmp = array[i]
    array[i] = array[min]
    array[min] = tmp
    print(array)
```

```python
#삽입정렬
array = [6,5,3,1,8,7,2,4]

for i in range(1,n):
    for j in range(i, 0, -1):
        if array[j-1] > array[j]:
            tmp = array[j]
            array[j] = array[j-1]
            array[j-1] = tmp
    print(array)
```
